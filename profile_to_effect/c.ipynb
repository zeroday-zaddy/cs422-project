```python
import pandas as pd
import numpy as np


pd.set_option('display.max_columns', None)  

data_df = pd.read_csv('../data/generated/data_matched_c.csv')
print(data_df.shape)
display(data_df.head(2))
```

```python
from sklearn import preprocessing
label_encoder_y = preprocessing.LabelEncoder()

data_df = data_df.dropna(subset=['Main_Effect'])
data_df = data_df.fillna(0)
print(data_df.shape)

print(data_df.columns)
X_cols = data_df.columns[9:]
y_col = data_df.columns[7]

X = data_df[X_cols].to_numpy(dtype=float)

label_encoder_y.fit(data_df[y_col])
y = label_encoder_y.transform(data_df[y_col])

print(label_encoder_y.classes_)
print(data_df.head(3))
```

```python
import matplotlib.pyplot as plt

data_df2 = data_df.copy(deep=True)
data_df2["Main_Effect"] = label_encoder_y.transform(data_df2["Main_Effect"])

data_df2.pivot(columns='Main_Effect',values='Main_Effect').plot.hist()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=777)

rf_model = RandomForestClassifier(random_state=777)

rf_model.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(rf_model.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(rf_model.score(X_test, y_test)))
```

```python
import mglearn
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), rf_model.estimators_)):
    ax.set_title("Tree {}".format(i))
    mglearn.plots.plot_tree_partition(X_train, y_train, tree, ax=ax)
    
mglearn.plots.plot_2d_separator(rf_model, X_train, fill=True, ax=axes[-1, -1],
                                alpha=.4)
axes[-1, -1].set_title("Random Forest")
mglearn.discrete_scatter(X_train[:, 0], X_train[:, 1], y_train)
```
